```python
import pandas as pd
import numpy as np
from pathlib import Path

RAW = Path("../data/raw")
print("Looking for data in:", RAW.resolve())
print("Files present:", sorted([f.name for f in RAW.glob("*.csv")]))
```

```python
# Load training data — takes about 30-60 seconds, uses ~2GB RAM
train_txn = pd.read_csv(RAW / "train_transaction.csv")
train_id = pd.read_csv(RAW / "train_identity.csv")

print(f"Transactions shape: {train_txn.shape}")
print(f"Identity shape:     {train_id.shape}")
print(f"Memory usage:       {train_txn.memory_usage(deep=True).sum() / 1e9:.2f} GB")
```

```python
# Quick look at what columns we have
print("=== Transaction columns (first 30) ===")
print(train_txn.columns[:30].tolist())
print(f"\n... and {len(train_txn.columns) - 30} more columns")

print("\n=== Identity columns ===")
print(train_id.columns.tolist())

print("\n=== Sample transactions ===")
train_txn.head(5)
```
